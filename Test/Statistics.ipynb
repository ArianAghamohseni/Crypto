Statistics question one

```python
def salary_stats(value):
    if value < 100000000000:
        return "low"
    if 10000 <= value < 300000000000:
        return "average"
    elif 25000 <= value < 500000000000:
        return "high"

q1 = df[['Name','timestamp','marketCap', 'volume']]
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(figsize = (8,8))

plt.scatter(q1['marketCap'], q1['volume'])

plt.title('Daily volume by market cap', fontsize = 14)
plt.xlabel('Coin market cap', fontsize = 10)
plt.ylabel('Coin daily volume', fontsize = 10)

plt.show()
```

```python
q1['mc_stats'] = q1['marketCap'].map(salary_stats)
q1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.lmplot(x = 'marketCap', y = 'volume', data = q1, hue = 'mc_stats', palette=['red','green','orange'], fit_reg = False)

plt.title('Daily volume by market cap Classified!', fontsize = 12)
plt.xlabel('Coin market cap', fontsize = 10)
plt.ylabel('Coin daily volume', fontsize = 10)

plt.show()
```
