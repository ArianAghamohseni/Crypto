```python
import sqlalchemy as db
from sqlalchemy.ext.declarative import declarative_base
import pandas as pd
```

```python
Username = 'root'
Pass = 'arvin1381'
Server = 'localhost'
engine = db.create_engine(f'mysql+pymysql://{Username}:{Pass}@{Server}:3306/queraprojectdb', echo = True)

conn = engine.connect()
Base = declarative_base()
```

```python
CoinTag = db.Table('CoinTag', Base.metadata,
                 db.Column('Name', db.String(255), db.ForeignKey('coins.name')),
                 db.Column('Tags', db.String(255), db.ForeignKey('tags.Tags')))

class Coin(Base):
    __tablename__ = 'coins'

    rank = db.Column(db.Integer)
    name = db.Column(db.String(255), primary_key=True)
    symbol = db.Column(db.String(255))
    main_link = db.Column('MainLink',db.String(255))
    historical_link = db.Column('HistoricalLink',db.String(255))
    github_link = db.Column('Github_link',db.String(255))

    tags = db.orm.relationship('Tag', secondary=CoinTag, back_populates='coins')
    coinhistories = db.orm.relationship('CoinHistory', back_populates='coins')

class Tag(Base):
    __tablename__ = 'tags'

    id = db.Column(db.Integer, autoincrement='auto')
    Tags = db.Column(db.String(255),primary_key=True)

    tags = db.orm.relationship('Coin', secondary=CoinTag, back_populates='tags')

class CoinHistory(Base):
    __tablename__ = "coinhistories"

    id = db.Column(db.Integer, primary_key=True, autoincrement='auto')
    name = db.Column(db.String(255), nullable=False)
    Date = db.Column(db.Date)
    open = db.Column(db.Float)
    timeHigh = db.Column(db.Time)
    high = db.Column(db.Float)
    timeLow = db.Column(db.Time)
    low = db.Column(db.Float)
    close = db.Column(db.Float)
    volume = db.Column(db.Float, nullable=False)
    marketCap = db.Column(db.Float, nullable=False)
    DailyPrice = db.Column(db.Float)
    CirculatingSupply = db.Column(db.Float)
    DailyChange = db.Column(db.Float)
    ChangeStatus = db.Column(db.String(255))

    coin = db.orm.relationship('Coin', back_populates='coinhistories')
```

```python
Base.metadata.create_all(engine)
```
